# I. Data Coding and Merging

```python
# Load packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load game logs data
Teams = pd.read_excel("../Data/Game logs 1999-2003.xlsx")
```

```python
#View DataFrame
Teams.head()
```

```python
# create dummy variable for home wins and create year variable
Teams['hwin']= np.where(Teams.home_score > Teams.visitor_score, 1, 0)
Teams['awin']= np.where(Teams.home_score < Teams.visitor_score, 1, 0)
Teams['year']= Teams.date.astype(str).str[0:4]
Teams.head()
```

```python
# Create Dataframe that sums stats you need by season 
Teamshome= Teams.groupby(['year', 'home'])['visitor_ab', 'visitor_h', 'home_h', 'home_ab', 'hwin', 'awin'].sum()
Teamshome= Teamshome.reset_index()
Teamshome= Teamshome.rename(columns={'home': 'Team'})
Teamshome.head()
```

```python
Teamsaway= Teams.groupby(['year', 'visitor'])['visitor_ab', 'visitor_h', 'home_h', 'home_ab', 'hwin', 'awin'].sum()
Teamsaway = Teamsaway.reset_index()
Teamsaway = Teamsaway.rename(columns = {"visitor": "Team"}) 
Teamsaway.head(10)
```

```python
#merge 2 dataframes on year and TEam
Teams2= pd.merge(Teamshome, Teamsaway, on=['Team', 'year'])
Teams2.head()
```

```python
Teams2['win'] = Teams2['hwin_x']+ Teams2['awin_y']
Teams2.head()
```

# II. Calculating Performance Metrics

```python
# create batting average for and batting average against as the sum the relevant statistics for each team
Teams2['AvgFOR']= (Teams2.home_h_x+ Teams2.visitor_h_y)/ (Teams2.home_ab_x + Teams2.visitor_ab_y)
Teams2['AvgAGN']= (Teams2.home_h_y+ Teams2.visitor_h_x)/ (Teams2.home_ab_y + Teams2.visitor_ab_x)
Teams2.head()
```

```python
# create win% by dividing games played and dividing total wins 
Teams3 = Teams2[['year', 'Team', 'win', 'AvgFOR', 'AvgAGN']]
Teams3.head()
```

```python
# get number of games played
TeamsGh= Teams.groupby(['year', 'home'])['hwin'].count()
TeamsGh= TeamsGh.reset_index()
TeamsGh= TeamsGh.rename(columns={'home': 'Team'})

TeamsGa= Teams.groupby(['year', 'visitor'])['awin'].count()
TeamsGa= TeamsGa.reset_index()
TeamsGa= TeamsGa.rename(columns={'visitor': 'Team'})

TeamsG= pd.merge(TeamsGh, TeamsGa, on=['Team', 'year'])
TeamsG['Games']= TeamsG.hwin + TeamsG.awin
TeamsG= TeamsG[['Team', 'year', 'Games']]
TeamsG.head(11)
```

```python
# merge datasets
Teams3= pd.merge(Teams3, TeamsG, on=['Team', 'year'])
Teams3.head()
```

```python
# create wpc for each team
Teams3['wpc']= Teams3.win/ Teams3.Games
Teams3.head()
```

```python
Teams3['AVG_dif']= Teams3.AvgFOR - Teams3.AvgAGN
Teams3.Games.value_counts()
```

# III. Running Regressions

```python
import statsmodels.formula.api as smf
#Run an OLS regression of win % on batting avg for and batting avg against 
reg1= smf.ols(formula='wpc ~ AvgFOR + AvgAGN', data=Teams3).fit()
reg1.summary()
```

```python
import statsmodels.formula.api as smf
#Run an OLS regression of win % on batting avg for and batting avg against 
reg2= smf.ols(formula='wpc ~ I(AvgFOR - AvgAGN)', data=Teams3).fit()
reg2.summary()
```
