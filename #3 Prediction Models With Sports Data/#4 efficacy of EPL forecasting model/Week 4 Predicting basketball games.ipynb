```python
import pandas as pd
import numpy as np
```

```python
NBAmod = pd.read_excel("Assignment Data/Week 4/NBA prediction model (Assignment).xlsx")
df=NBAmod
df.head()
```

```python
# define probabilites assoicated with bookmarker odds
df['bookieprobH']= 1/ df.hwinodds / (1/df.hwinodds + 1/df.hloseodds)
df['bookieprobA']= 1/ df.hloseodds / (1/df.hwinodds + 1/df.hloseodds)

# the variable 'win' gives us value of 1 when the home team wins
# create variable lose which 1 when home team loses 
df['hlose']= np.where(df.hwin== 0, 1, 0)

df.head(10)
```

```python
# define the predicted result based on the betting odds 
df['bookres']= np.where(df.bookieprobH >df.bookieprobA, 'H', 'A')
df['FTR']= np.where(df.hwin ==1, 'H', 'A')
df.head()
```

```python
# cross tabl showing how often the bookmarker odds correctly predict the result
pd.crosstab(df.FTR, df.bookres, dropna=True)
```

```python
# define variable equal to log of the ratio of the home salary 
# create log of salary ratios
lg_home= np.log(df.hteamsal)
lg_away= np.log(df.opposal)
df['lg_sal_ratio']= lg_home- lg_away
df.head()
```

```python
# define a subset of cal year 2018 data
df_train= df[df.year== 2018]
df_train.head()
```

```python
# import logisitc regression
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# run the logistic regression of hwin on the log salary ratio
H_win_lg= 'hwin ~ lg_sal_ratio'
model= smf.glm(formula=H_win_lg, data=df_train, family=sm.families.Binomial())
result=model.fit()

print(result.summary())
```

```python
# predicted proabailty of home win
df['predprobH']= 1/(1+1/(np.exp(.4452+1.1216*df['lg_sal_ratio'])))
df.head()
```

```python
# if greater .5 its a win
df['predres']= np.where(df.predprobH > .5, 'H', 'A')
df
```

```python
# create for 2019 test
df2019= df[df.year == 2019]
df2019['booktrue']= np.where(df2019.FTR == df2019.bookres, 1, 0)
df2019['modeltrue']= np.where(df2019.FTR == df2019.predres, 1,0)
df2019
```

```python
# accuracy of bookie vs model
print('bookie:', df2019.booktrue.mean())
print('model:', df2019.modeltrue.mean())
```

```python
# get brier score of both (lower brier the better the model is)
df2019['Brierbookie']= (df2019.hwin - df2019.bookieprobH) ** 2 + (df2019.hlose - df2019.bookieprobA) ** 2
print('brier score bookie:', df2019.Brierbookie.mean())
df2019['Briermodel']= (df2019.hwin - df2019.predprobH) ** 2 + (df2019.hlose - df2019.predprobH) ** 2
print('brier score model:', df2019.Briermodel.mean())
```
