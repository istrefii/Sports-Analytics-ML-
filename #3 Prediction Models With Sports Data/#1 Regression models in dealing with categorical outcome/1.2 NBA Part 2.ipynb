### We will use the latter half of the 2017 regular season data as a D.V for forecasting

```python
# Note: All the pre-season games are recorded with GAME_ID larger than 1000000000.  \n",
NBA17 = NBA[(NBA.SEASON_ID == 22017) & (NBA.GAME_ID < 1000000000)]
display(NBA17[0:10])
```

```python
# Display columns of the data
NBA17.columns
```

```python
# Select columns used for forecasting model: 'GAME_ID', 'TEAM_NAME', 'GAME_DATE','PTS','PLUS_MINUS', 'WIN'\n",
NBA17 = NBA17[['GAME_ID', 'TEAM_NAME', 'GAME_DATE','PTS','PLUS_MINUS', 'WIN']]
display(NBA17[0:10])
```

```python
# subset the data for forecasting
# tip: you can roughly split the data half and half  by using the GAME ID column (the balance point: GAME ID == 21700615)

NBA17_pre= NBA17[NBA17.GAME_ID <= 21700615].copy()
# Check the output 
NBA17_pre.sort_values('GAME_ID', axis=0, ascending=True, inplace=True, na_position='last')
NBA17_pre.head()

```

```python
# Create test
NBA17_post= NBA17[NBA17.GAME_ID > 21700615].copy()
NBA17_post.sort_values('GAME_ID', axis=0, ascending=True, inplace=True, na_position='last')
NBA17_post.head()
```

```python
# create a new column named 'PTS_AGN for creating the Pyth %

NBA17_pre['PTS_AGN']= NBA17_pre['PTS']- NBA17_pre.PLUS_MINUS
NBA17_pre.head()
```

```python
# Aggregate game level statistics to a team level statistics for forecasting
NBA17_pre_team= NBA17_pre.groupby(['TEAM_NAME'])['WIN', 'PTS', 'PTS_AGN'].sum()
NBA17_pre_team.head()
```

```python
# create new column containing the number of games played for the first half of reg season
## Note: While the balance point (21700615) was used to split data in half, number of games played by each team vaires slightly from each other
## So, we cant use 41 (82/2) for calcuatling win % and it necessart to create df containng the exact number of games played in teh first half by each team
games_count= NBA17_pre.groupby(['TEAM_NAME']).size().reset_index(name= 'GAMES_PLAYED')
games_count.head()
```

###### As you can see the resulting df, there is slight variation from the center (41) among teams 

```python
# merge games_count into nba17_pre_team dataset
NBA17_pre_team= pd.merge(NBA17_pre_team, games_count, on='TEAM_NAME')
NBA17_pre_team.head()
```

```python
# create the 1) Pyhtagorean win % and 2) the win % for the basis of forecasting the remainder of the game 
NBA17_pre_team['wpc_pre'] = NBA17_pre_team.WIN/ NBA17_pre_team.GAMES_PLAYED
NBA17_pre_team['pyth_pre'] = NBA17_pre_team.PTS ** 2 / (NBA17_pre_team['PTS'] **2 + NBA17_pre_team['PTS_AGN']**2)
NBA17_pre_team.head()
```

```python
sns.relplot(x='pyth_pre', y='wpc_pre', data=NBA17_pre_team)
```

```python
# correlation between Pyth win % and win % for the 1st half of the regular season
NBA17_pre_team['pyth_pre'].corr(NBA17_pre_team.wpc_pre)
```

```python
# drop unncessary columns
NBA17_pre_team.drop(['WIN', 'PTS', 'PTS_AGN', 'GAMES_PLAYED'], axis=1, inplace=True)
```

### we will use the latter half of the 2017 regular seaosn datat as DV for forecasting

```python
NBA17_post.head()
```

```python
# aggergate game level stats to a team level for forecasting
NBA17_post_team = NBA17_post.groupby(['TEAM_NAME'])['WIN', 'PTS'].sum()
NBA17_post_team.head()
```

```python
NBA17_games_played= NBA17_post.groupby(['TEAM_NAME']).size().reset_index(name='GAMES_PLAYED')
NBA17_games_played
```

```python
NBA17_post_team= pd.merge(NBA17_post_team, NBA17_games_played, on='TEAM_NAME')
NBA17_post_team.head()
```

```python
NBA17_post_team['wpc_post']= NBA17_post_team['WIN']/NBA17_post_team.GAMES_PLAYED
NBA17_post_team.head()
```

```python
# drop unncessary columns
NBA17_post_team.drop(['WIN', 'PTS', 'GAMES_PLAYED'], axis=1, inplace=True)
NBA17_post_team.head()
```

 ### Merge NBA17_pre_team data to NBA17_post_team data

```python
NBA17_post_team= pd.merge(NBA17_post_team, NBA17_pre_team, on='TEAM_NAME')
NBA17_post_team.head()
```

```python
# lets plot the variables and fit the model using wpc_pre as I.V
sns.regplot(x='wpc_pre', y='wpc_post', data=NBA17_post_team, marker='.')
plt.xlabel('1st half win%')
plt.ylabel('2nd half win%')
plt.title('Correlation: Pre-win% vs Post-win%', fontsize=15)
```

```python
WinPct_lm= smf.ols(formula = 'wpc_post ~ wpc_pre', data=NBA17_post_team).fit()
WinPct_lm.summary()
```

```python
# lets plot the variables and fit the model using pyth as I.V
sns.regplot(x='pyth_pre', y='wpc_post', data=NBA17_post_team, marker='.')
plt.xlabel('1st half win%')
plt.ylabel('2nd half win%')
plt.title('Correlation: Pyth-win% vs Post-win%', fontsize=15)
```

```python
PythWin_lm= smf.ols(formula = 'wpc_post ~ pyth_pre', data=NBA17_post_team).fit()
PythWin_lm.summary()
```

```python
from statsmodels.iolib.summary2 import summary_col
info_dict= {'R-squared': lambda x: x.rsquared}
Header= ['1', '2']
Table_1= summary_col([WinPct_lm, PythWin_lm], regressor_order=['Intercept', 'pyth_pre', 'wpc_pre'], model_names=Header)
print(Table_1)
```
