## Part 2 - Regression Analysis 1 - Test of Home Game Advantage

```python
reg1= sm.ols(formula='score_diff ~ home', data=df).fit()
print(reg1.summary())
```

```python
Reg1_2 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral', data= df).fit()
print(Reg1_2.summary())
```

```python
Reg1_3 = sm.ols(formula = 'score_diff ~ home+stadium_capacity+stadium_neutral+home*stadium_neutral+team+opponent', data= df).fit()
print(Reg1_3.summary())
```

## Part 3 - Regression Analysis 2 -- Impact of Outside Factors on Scores

```python
reg2_1= sm.ols(formula='score ~ season +home', data=df).fit()
print(reg2_1.summary())
```

```python
reg2_2= sm.ols(formula='score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity', data=df).fit()
print(reg2_2.summary())
```

```python
reg2_3= Reg1_2 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity+ stadium_capacity+stadium_age+stadium_type+stadium_neutral+ home*stadium_neutral', data= df).fit()
print(reg2_3.summary())
```

```python
reg2_4 = sm.ols(formula = 'score ~ season+home+weather_temperature+weather_wind_mph+weather_humidity+ stadium_capacity+stadium_age+stadium_type+stadium_neutral+ home*stadium_neutral+team+opponent', data= df).fit()
print(reg2_4.summary())
```
