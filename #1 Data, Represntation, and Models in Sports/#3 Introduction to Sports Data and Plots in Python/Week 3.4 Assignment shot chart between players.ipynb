## Part 1 - NBA Player Comparision

```python
# As usual, we begin by importing the packages we will need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The data consists of the shot log for the NBA season 2016/17

shot = pd.read_csv("Assignment Data/Week 3/NBA shotlog_16_17.csv")
pd.set_option('display.max_columns', 100)
print(shot.columns.tolist())
shot.head()
```

```python
# get all shots onto a half court by defining shot
shot['halfcourt_x'] =np.where(shot['location_x'] < 933/2, 933 - shot['location_x'],shot['location_x'])
shot['halfcourt_y'] =np.where(shot['location_x'] < 933/2, 500 - shot['location_y'],shot['location_y'])
shot.head()
```

```python
# define subset for diff player
rus= shot[shot.shoot_player == 'Russell Westbrook']
durant= shot[shot.shoot_player == 'Kevin Durant']
dwight= shot[shot.shoot_player == 'Dwight Howard']
jordan= shot[shot.shoot_player== 'DeAndre Jordan']
brook= shot[shot.shoot_player== 'Brook Lopez']
robin= shot[shot.shoot_player== 'Robin Lopez']
```

```python
hxrw= rus['halfcourt_x']
hyrw= rus['halfcourt_y']
hxkd= durant['halfcourt_x']
hykd= durant['halfcourt_y']
hxdw= dwight['halfcourt_x']
hydw= dwight['halfcourt_y']
hxdj= jordan['halfcourt_x']
hydj= jordan['halfcourt_y']
hxbl= brook['halfcourt_x']
hybl= brook['halfcourt_y']
hxrl= robin['halfcourt_x']
hyrl= robin['halfcourt_y']
```

## Westbrook - Durant Comparision

```python
f = plt.figure(figsize=(94/6,50/6))
ax = f.add_subplot(121)
colors = np.where(rus['current_shot_outcome']=='SCORED','g',np.where(rus['current_shot_outcome']=='MISSED','r','b'))
ax = plt.scatter(hxrw,hyrw, s=10, c= colors, marker= '.')
plt.grid(True)
plt.xlim(500,950)
plt.ylim(0,500)
plt.title("Russell Westbrook", fontsize = 15)
ax = f.add_subplot(122)
colors = np.where(durant['current_shot_outcome']=='SCORED','g',np.where(durant['current_shot_outcome']=='MISSED','r','b'))
ax = plt.scatter(hxkd,hykd, s=10, c= colors, marker= '.')
plt.grid(True)
plt.xlim(500,950)
plt.ylim(0,500)
plt.title("Kevin Durant", fontsize = 15)
```
