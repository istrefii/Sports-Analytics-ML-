### Step 1


```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
import matplotlib.dates as mdates
import datetime
import plotly.graph_objects as go

import scipy.stats as stats

df = pd.read_csv("../Data/FH.csv")
df.head()
```

```python
df.drop(['Unnamed: 0', 'Latitude', 'Longitude', 'Heart Rate'], axis=1, inplace=True)
```

### Step 2

```python
# create a copy of df and clean up columns consider multi-index
master= df.set_index(['Timestamp', 'AthleteID'], inplace=False)
master.head()
```

### Step 3

```python
# make new variable 20 seconds furtherest distance and 60 distance

master['farthest']= master.Odometer.diff(200)

master['one_mindistance']= master.Odometer.diff(600)

master.loc[master.one_mindistance < 0, 'one_mindistance'] = np.nan
master.loc[master.farthest < 0, 'farthest']= np.nan

master.head()
```

### Step 4

```python
max_value= master.groupby('AthleteID').agg([max])
max_value
```

### Step 5

```python
# find z score
max_value['zscores_vel']= stats.zscore(max_value.Velocity)
max_value['zscores_far']= stats.zscore(max_value.farthest)
max_value['zscores_one']= stats.zscore(max_value.one_mindistance)
max_value
```

### Step 6

```python
# adding up the z-scores to provide my simple global metric! 

max_value['Metric'] = stats.zscore(max_value['Velocity']) + stats.zscore(max_value['farthest'])+stats.zscore(max_value['one_mindistance']) + 10

print(max_value)
```
